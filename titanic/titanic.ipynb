导入需要的库

```python
import pandas as pd
import numpy as np
from sklearn.neural_network import MLPClassifier
```

读取csv中的数据

```python
df = pd.read_csv('train.csv')
df
```

检查包含空值的数据列

```python
df.isnull().any()
```

先处理Age列的空值，这里用年龄均值填充Nan数据

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

接下来处理Embarked列的数据，作为登船港口的数据，这里用众数取代Nan值

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df
```

检查是否有空数据

```python
df.isnull().any()
```

为了训练方便，把Sex和Embarked的值以数值替代

```python
df.loc[df['Embarked'] == 'S', 'Embarked'] = 0
df.loc[df['Embarked'] == 'C', 'Embarked'] = 1
df.loc[df['Embarked'] == 'Q', 'Embarked'] = 2
df.loc[df['Sex'] == 'male', 'Sex'] = 1
df.loc[df['Sex'] == 'female', 'Sex'] = 2
df
```

把数据填入list做进一步处理

```python
df_list = []
df_list.append(df)
```

筛选可以用于训练的数据，并整合到数组中作为训练的输入和输出

```python
input_cols = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
output_col = 'Survived'

x_tr = []
y_tr = []

for data in df_list:
    x_tr.append(data[input_cols].values)
    y_tr.append(data[output_col].values)
x_train = np.concatenate(x_tr, axis=0)
y_train = np.concatenate(y_tr)
print(x_train)
print(y_train)
```

构建mlp模型并训练

```python
mlp = MLPClassifier(hidden_layer_sizes=64, alpha=0, max_iter=200, random_state=42, learning_rate_init=1e-3)
mlp.fit(x_train, y_train)
```

导入测试数据集

```python
test_df = pd.read_csv('test.csv')
test_df
```

```python
test_df.isnull().any()
```

```python
test_df['Age'].fillna(test_df['Age'].mean(), inplace=True)
test_df['Fare'].fillna(test_df['Fare'].mode()[0], inplace=True)
test_df.isnull().any()
```

用数值代替Sex和Embarked

```python
test_df.loc[test_df['Embarked'] == 'S', 'Embarked'] = 0
test_df.loc[test_df['Embarked'] == 'C', 'Embarked'] = 1
test_df.loc[test_df['Embarked'] == 'Q', 'Embarked'] = 2
test_df.loc[test_df['Sex'] == 'male', 'Sex'] = 1
test_df.loc[test_df['Sex'] == 'female', 'Sex'] = 2
test_df
```

```python
test_list = []
test_list.append(test_df)

# 筛选用于训练的数据列
test_input_cols = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

x_test = []

for data in test_list:
    x_test.append(data[test_input_cols].values)

x_test_set = np.concatenate(x_test, axis=0)
x_test_set.shape
```

```python
result = mlp.predict(x_test_set)
result
```

```python
test_df['Survived'] = result
test_df
```

导出csv

```python
test_df.to_csv(path_or_buf='predict_result.csv', columns=['PassengerId', 'Survived'], index=False)
```
